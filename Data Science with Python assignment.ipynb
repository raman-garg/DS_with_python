```python
#Numerical data
numerical_data=house_data.select_dtypes(['float64','int64']).columns
numerical_data
```

```python
#Categorical Data
categorical_data=house_data.select_dtypes(['object']).columns
categorical_data
```

```python
#Correcting Data types
for col in house_data.columns:
    if(house_data[col].dtype == 'object'):
        house_data[col] = house_data[col].astype('category')
```

```python
house_data.info()
```

```python
pd.isnull(house_data).sum()
```

```python
#filling null values
ist=round(house_data['Bedroom2'].mean())
print(ist)
```

```python
house_data['Bedroom2'].fillna(ist,inplace=True)
```

```python
bath=round(house_data['Bathroom'].mean())
house_data['Bathroom'].fillna(bath,inplace=True)
```

```python
land = round(house_data['Landsize'].mean())
house_data['Landsize'].fillna(land,inplace=True)
```

```python
#Outliers Treatment
import seaborn as sns
```

```python
sns.boxplot(house_data['Distance'])
```

```python
q1 = house_data['Distance'].quantile(0.25)
q1
```

```python
q3 = house_data['Distance'].quantile(0.75)
q3
```

```python
##IQR = q3 - q1
IQR = q3 - q1
IQR
```

```python
floor = 1.5 * IQR - q1
floor
```

```python
cap = q3 + 1.5 * IQR
cap
```

```python
house_data = house_data[(house_data['Distance'] >= floor) & (house_data['Distance'] <= cap)]
```

```python
len(house_data)
```

```python
house_data.describe()
```

```python
house_data.shape
```

```python
#Feature engineering
#Feature extraction
house_data.loc[house_data['Landsize'] >15000, 'Cost'] = 'High'
house_data.loc[(house_data['Landsize'] >1000)&(house_data['Landsize'] <=15000), 'Cost'] = 'Medium'
house_data.loc[house_data['Landsize'] <1000, 'Cost'] = 'Low'
```

```python
house_data.head()
```

```python
#Feature normalization
int_columns = house_data.select_dtypes(['int64','float64']).columns
int_columns.value_counts()
```

```python
##feature z-score
from scipy.stats import zscore
```

```python
house_int = house_data[int_columns].apply(zscore)
```

```python
house_out = house_int.drop(['Car','BuildingArea','Price','YearBuilt','Lattitude','Longtitude','Propertycount'], axis =1)
```
